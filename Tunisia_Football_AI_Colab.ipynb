## 4️⃣ Model Training & Fine-tuning

**Train YOLO on Tunisia-specific football data**

### 4.1 Prepare Training Data

**Upload annotated images in YOLO format:**
```
dataset/
├── images/
│   ├── train/
│   │   ├── img001.jpg
│   │   └── ...
│   └── val/
│       ├── img100.jpg
│       └── ...
└── labels/
    ├── train/
    │   ├── img001.txt
    │   └── ...
    └── val/
        ├── img100.txt
        └── ...
```

**Label format (one line per object):**
```
class_id center_x center_y width height
2 0.5 0.3 0.1 0.2  # player at center
0 0.8 0.9 0.05 0.05  # ball at bottom right
```

**Classes:**
- 0: ball
- 1: goalkeeper
- 2: player
- 3: referee

````python
# Option 1: Upload dataset as ZIP
from google.colab import files
import zipfile

print("📦 Upload your dataset ZIP file (YOLO format)")
uploaded = files.upload()

for filename in uploaded.keys():
    with zipfile.ZipFile(filename, 'r') as zip_ref:
        zip_ref.extractall('training_data/')
    print(f"✅ Extracted: {filename}")

# Option 2: Download from URL
# !wget https://your-dataset-url.com/tunisia_football_dataset.zip
# !unzip tunisia_football_dataset.zip -d training_data/
````

````python
# Create dataset.yaml configuration
dataset_yaml = """
path: /content/tunisia-football-ai/Field_Fusion/training_data
train: images/train
val: images/val

nc: 4  # number of classes
names: ['ball', 'goalkeeper', 'player', 'referee']
"""

with open('dataset.yaml', 'w') as f:
    f.write(dataset_yaml)

print("✅ dataset.yaml created")
!cat dataset.yaml
````

### 4.2 Fine-tune Existing Model (Recommended)

**Start from the HuggingFace football model and fine-tune on Tunisia data**

````python
from ultralytics import YOLO
from huggingface_hub import hf_hub_download

# Download base model
print("📥 Downloading base football model from HuggingFace...")
base_model_path = hf_hub_download(
    repo_id="uisikdag/yolo-v8-football-players-detection",
    filename="best.pt"
)

# Load model
model = YOLO(base_model_path)

# Fine-tune on Tunisia data
print("\n🔧 Fine-tuning on Tunisia-specific data...")
results = model.train(
    data='dataset.yaml',
    epochs=50,  # Adjust based on dataset size
    imgsz=640,
    batch=16,
    name='tunisia_football_v1',
    patience=10,  # Early stopping
    save=True,
    device=0,  # GPU
    workers=2,
    plots=True
)

print("\n✅ Training complete!")
print(f"   Best model saved to: runs/detect/tunisia_football_v1/weights/best.pt")
````

### 4.3 Train from Scratch (Advanced)

**Train a new YOLOv8 model from scratch**

````python
from ultralytics import YOLO

# Load YOLOv8 base model
model = YOLO('yolov8n.pt')  # Use 'yolov8s.pt' or 'yolov8m.pt' for better accuracy

# Train from scratch
print("🏋️ Training from scratch...")
results = model.train(
    data='dataset.yaml',
    epochs=100,
    imgsz=640,
    batch=16,
    name='tunisia_football_scratch',
    patience=20,
    save=True,
    device=0,
    workers=2,
    plots=True,
    lr0=0.01,  # Initial learning rate
    lrf=0.01,  # Final learning rate
    momentum=0.937,
    weight_decay=0.0005
)

print("\n✅ Training complete!")
````

### 4.4 Evaluate Model Performance

````python
# Load trained model
model = YOLO('runs/detect/tunisia_football_v1/weights/best.pt')

# Evaluate on validation set
metrics = model.val()

print("\n📊 Model Performance:")
print(f"   mAP@0.5: {metrics.box.map50:.3f}")
print(f"   mAP@0.5:0.95: {metrics.box.map:.3f}")
print(f"   Precision: {metrics.box.mp:.3f}")
print(f"   Recall: {metrics.box.mr:.3f}")

# Show confusion matrix
from IPython.display import Image, display
display(Image('runs/detect/tunisia_football_v1/confusion_matrix.png'))
````

### 4.5 Test Model on New Video

## 💡 Tips & Best Practices

### **For Best Results:**

1. **Dataset Quality**
   - Minimum 500-1000 images per class
   - Diverse lighting conditions
   - Different camera angles
   - Tunisia-specific stadiums & jerseys

2. **Training Tips**
   - Start with fine-tuning (faster, better results)
   - Use data augmentation (built-in YOLO)
   - Monitor validation loss (early stopping)
   - Save checkpoints every 10 epochs

3. **GPU Usage**
   - Use Colab Pro for longer training times
   - Enable GPU: Runtime → Change runtime type → GPU
   - Monitor GPU memory: `!nvidia-smi`

4. **Model Deployment**
   - Export to ONNX for CPU deployment
   - Use TensorRT for NVIDIA GPUs
   - Test on multiple videos before production

### **Annotation Tools:**
- **Roboflow**: https://roboflow.com (Auto-annotation)
- **LabelImg**: Desktop tool for YOLO format
- **CVAT**: https://cvat.org (Team collaboration)
